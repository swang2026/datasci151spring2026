##### to add a column to a dataset --> dataset[column] = column_data

##### use .apply() to apply a function onto all the data in a dataset

```python
emissions = pd.read_csv("practice_quiz/quiz_data/china_emission.csv")

def minus_10(x):
    return x - 10

energy_minus_10 = emissions["Energy_consumption"].apply(minus_10)

emissions["Energy_minus_10"] = energy_minus_10
print(emissions.head())
```

##### use .shape to find the (rows,column) shape of a dataset

```python
num_rows, num_cols = emissions.shape
print(num_cols, num_cols)
```

##### use .query("logical expression") to only find data points with that requirement

#####

```python
high_emissions_over6000 = emissions.query("(CO2_emission >= 6000) and (GDP > 20000)")
print(high_emissions_over6000.head())
```
