##### use .apply() to apply a function onto all the data in a dataset

```python
emissions = pd.read_csv("practice_quiz/quiz_data/china_emission.csv")

def minus_10(x):
    return x - 10

energy_minus_10 = emissions["Energy_consumption"].apply(minus_10)

emissions["Energy_minus_10"] = energy_minus_10
print(emissions.head())
```

##### use .shape to find the (rows,column) shape of a dataset

##### use .query("logical expression") to only find data points with that requirement

##### use `` if the logical expression has spaces in it (only used on name)

##### use @ if before the global variable if used in the logical expression

```python
high_emissions_over6000 = emissions.query("(CO2_emission >= 6000) and (GDP > 20000)")
print(high_emissions_over6000.head())

high_emission = 6000
high_emissions_over6000_new = emissions.query("CO2_emission >= @high_emission")
display(high_emissions_over6000_new.head())
```

##### different ways to change one index

```python
student_roster = ["Nancy", "Bill", "Elaine", "Boris", "Assata"]
student_grades = [92.8, 87.3, 95.4, 82.3, 98.7]
b_index = student_roster.index("Boris") # find index
student_grades[b_index] = 85 # update corresponding element
print(student_grades)

#dictionary example
grades = {"Nancy" : 92.8, "Bill" : 87.3, "Elaine" : 95.4, "Boris": 82.3, "Assata" : 98.7}
grades["Boris"] = 85
print(grades["Boris"])
```

##### these dictionaries have keys (names) and values

```python
print(grades.keys())
print(grades.values())
```
