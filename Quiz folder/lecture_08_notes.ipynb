## For Loops

```python
friend_list = ["Susanna", "Abby", "Michelle", "Angie"]

for friends in friend_list:
    print("I am friends with " + friends + "!")
```

### For Loops using indexes

```python
friend_list = ["Susanna", "Abby", "Michelle", "Angie"]
index = 1

for friends in friend_list:
    print(friends + " is my number " + str(index) + " friend!")
    index = index + 1
```

### For Loops using range(N)

```python
friend_list = ["Susanna", "Abby", "Michelle", "Angie"]

N = len(friend_list)
print(range(N))
print(list(range(N))) # turns range into a list
print()

for index in range(N):
    print(friend_list[index] + " is my number " + str(index + 1) + " friend!")
```

### Interrupting For Loops
##### breaks - ends all the loops it is in
##### continue - skips everything after it and moves onto the next value

```python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2]

list_sum = 0
for value in list_mixed:
    if (type(value) != int) and (type(value) != float):
        print("Stopping sum computation: Non-numerical element in list!")
        break #although it is only in the if statement, it breaks out of the for loop and goes to the next command

    list_sum = list_sum + value

print(list_sum)
```

```python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2, None, "missing", 3]

list_sum = 0
for value in list_mixed:
    if (type(value) != int) and (type(value) != float):
        print("Warning: Non-numerical element in list!")
        continue #skip everything after this and go back to try next value in for loop

    list_sum = list_sum + value

print(list_sum)
```

## While Loops

```python
eps = 0.025

x_list = []
y_list = []

x = 1
while 1/x > eps:
    x_list.append(x)
    y_list.append(1 / x)
    x = x + 1

plt.scatter(x_list, y_list)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x) = 1 / x")
plt.show()
```

## Zip functions

##### If the different lists are different sizes, it will just go up to the shortest one

```python
movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]
movie_years = ["1964", "2022", "1960"]
movie_directors = ["Jacques Demy", "Gerard Johnstone", "Alfred Hitchcock"]

for title, rating, year, director in zip(movie_titles, movie_ratings, movie_years, movie_directors):
    message = "'" + title + "' (" + year + ") was directed by " + director + ". IMDB rating: " + str(rating)
    print(message)
```

```python
friend_names = ["Angie", "Abby", "Michelle", "Gabby"]
friend_color = ["green", "pink", "purple", "light pink"]

for name, color in zip(friend_names, friend_color):
    print(name + "'s favorite color is " + color + "!")
```
