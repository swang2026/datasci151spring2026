```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house = pd.read_csv("data/house_proportions.csv")
senate = pd.read_csv("data/senate_proportions.csv")
pres = pd.read_csv("data/presidential_proportions.csv")

display(house)
display(senate)
display(pres)
```

```python
plt.scatter(pres["Dem-Rep"], house["Proportion_Democrat"], label = "Dem", color = "blue")
plt.scatter(pres["Dem-Rep"], house["Proportion_Republican"], label = "Rep", color = "red")
plt.show()

plt.scatter(pres["Dem-Rep"], senate["Dem Proportion"], label = "Dem", color = "blue")
plt.scatter(pres["Dem-Rep"], senate["Rep Proportion"], label = "Rep", color = "red")
plt.show()



```
