```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house = pd.read_csv("data/house_proportions.csv")
senate = pd.read_csv("data/senate_proportions.csv")
pres = pd.read_csv("data/presidential_proportions.csv")

senate = senate.fillna(0)

display(house)
display(senate)
display(pres)
```

```python
plt.scatter(pres["Dem-Rep"], house["Proportion_Democrat"], label = "Dem", color = "blue")
plt.scatter(pres["Dem-Rep"], house["Proportion_Republican"], label = "Rep", color = "red")
plt.xlim(-0.5, 0.5)
plt.xlabel("Dem-Rep Proportion in Presidential Election")
plt.ylabel("Proportion of Votes in House Election")
plt.title("Proportion of Votes in House vs Presidential Election")
plt.legend()
plt.show()

senate_list = senate["state"].tolist()
pres_sen = pres.query("State in @senate_list")

plt.scatter(pres_sen["Dem-Rep"], senate["Dem Proportion"], label = "Dem", color = "blue")
plt.scatter(pres_sen["Dem-Rep"], senate["Rep Proportion"], label = "Rep", color = "red")
plt.xlabel("Dem-Rep Proportion in Presidential Election")
plt.ylabel("Proportion of Votes in Senate Election")
plt.title("Proportion of Votes in Senate vs Presidential Election")
plt.legend()
plt.show()
```
