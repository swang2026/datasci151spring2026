### For presidential party votes with a larger margin of victory are there more independent votes?

```python
import pandas as pd
```

```python
pres = pd.read_csv("data/president_county_candidate.csv")

state_total = pd.read_csv("data/president_state.csv")

county = pd.read_csv("data/president_state.csv")
```

```python
display(pres)

display(state_total)
```

```python
new_list = pres[["state", "party", "total_votes"]]

new_pres = pd.merge(left = new_list, right = state_total, how = "left", on = "state")

display(new_pres)
```

```python
bob = new_pres.groupby( ["state", "party"] ).agg(state_vote = ('total_votes_x', 'sum'))
bob = bob.reset_index()
print(bob)
```

```python
bob2 = pd.merge(left = bob, right = state_total, how = "left", on = "state")

bob2
```

```python
bob2["percentage"] = bob2["state_vote"] / bob2["total_votes"] * 100

display(bob2)
```

```python
find_ind = bob2.loc[bob2["party"] == "IND"]

find_ind
```
